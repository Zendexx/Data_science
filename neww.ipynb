```python
import numpy as np



a =np.array(6)
b =np.array([2,6,8,5])
c =np.array([[2,4,6],[6,6,7]])
d =np.array([[[4,6,7],[4,6,8]],[[5,7,8],[6,7,8]]])

print("0D:" ,a)
print("1D:" ,b)
print("2D:" ,c)
print("3D:" ,d)

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
print(b[2])
print(c[1,2])

print("sum:",b[2] + b[3])
```

```python
import numpy as np
import matplotlib.pyplot as plt
x =np.array([0,1,2,3,4])
y =np.array([0,1,2,3,4])
plt.plot(x,y,marker = "+")
plt.title("My Graph")
plt.xlabel("x -axis")
plt.ylabel("y -axis")

plt.grid()
plt.show()
```

```python
x_axis =np.array([0,3,5,7])
y_axis =np.array([0,5,8,16])

plt.scatter(x_axis, y_axis)
plt.show()
```

```python
x =np.array(["apples","Banana","panda","goat","sheep"])
y =np.array([4,5,2,9,7])

plt.bar(x,y, color="#080808",width=0.4)
plt.show()
```

```python
x =np.array(["apples","Banana","panda","goat","sheep"])
y =np.array([4,5,2,9,7])

plt.barh(x,y, color="blueviolet", height=0.5)
plt.show()
```

```python
x =np.array([70,50,60,90,10])
my_label =["panda","goat","donkey","sheep","cow"]
my_colors =["blueviolet","pink","black","lightblue","yellow"]
my_explode =[0.3,0,0,0,0]

plt.pie(x, labels=my_label, colors=my_colors, explode=my_explode , shadow=True)
plt.legend(title="Genus")
plt.show()
```
