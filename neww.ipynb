```python
import numpy as np



a =np.array(6)
b =np.array([2,6,8,5])
c =np.array([[2,4,6],[6,6,7]])
d =np.array([[[4,6,7],[4,6,8]],[[5,7,8],[6,7,8]]])

print("0D:" ,a)
print("1D:" ,b)
print("2D:" ,c)
print("3D:" ,d)

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
print(b[2])
print(c[1,2])

print("sum:",b[2] + b[3])
```
