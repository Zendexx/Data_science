```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random

x =random.randint(100,size=(5))
y =random.randint(100,size=(5))
plt.plot(x,y,marker = "+")
plt.title("My Graph")
plt.xlabel("x -axis")
plt.ylabel("y -axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random

x =random.randint(50,size=(5))
y =random.randint(50,size=(5))
plt.bar(x,y, color="#080808",width=1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random

x =random.randint(180,size=(5))
my_label =["panda","goat","donkey","sheep","cow"]
my_colors =["blueviolet","pink","black","lightblue","yellow"]

plt.pie(x, labels=my_label, colors=my_colors, shadow=True)
plt.legend(title="Genus")
plt.show()
```
